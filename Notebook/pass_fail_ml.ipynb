# Pass-fail model

### Importing Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

### Reading CSV

```python
student_df = pd.read_csv(r"C:\Users\Ire\Desktop\student_passfail_ml\data\student_ml_data.csv")

print(student_df)
```

### Subsetting

```python

a = student_df[['avg_grade', 'attendance_rate', 'num_courses', 'total_units']]


b = student_df['pass_fail']
```

### Training

```python
a_train, a_test, b_train, b_test = train_test_split(
    a, b, test_size=0.2, random_state=42
)
```

### Regression Training

```python
model = LogisticRegression()
model.fit(a_train, b_train)
```

### Predictions

```python
b_pred = model.predict(a_test)
```

### Evaluation

```python

print(f"Accuracy: {accuracy_score(b_test, b_pred)}")


print(f"Confusion Matrix:\n {confusion_matrix(b_test, b_pred)}")


print(f"\nClassification Report:\n {classification_report(b_test, b_pred)}")
```

Because dataset is very small this accuracy(1.0) is not generalizable. Even still 100% is 100%

### Coefficients Inspection

```python
coefficients = pd.DataFrame({
    'Feature': a.columns,
    'Coefficient': model.coef_[0]
})
print(coefficients)
```
